```python
# I've check the amazon for these missing data but I found nothing so we will drop these
df_cleaned = df.dropna()
```

```python
df_cleaned.shape
```

```python
df_cleaned.isnull().sum()
```

```python
# Next, we will add a new 'category' column because the category column is too long

df_cleaned['new_category'] = df_cleaned['category'].str.split(pat='|').str[-1]
```

```python
df_cleaned['new_category']
```

```python
df_cleaned.tail()
```

```python
#Next, we will change the data types of discounted_price, actual_price, discount_percentage, rating, rating_count

df_cleaned['rating_count'] = df_cleaned['rating_count'].str.replace(',', '').astype('int')
```

```python
df_cleaned['discounted_price'] = df_cleaned['discounted_price'].str.replace('‚Çπ', '').str.replace(',', '').astype('float')
```

```python
df_cleaned['actual_price'] = df_cleaned['actual_price'].str.replace('‚Çπ', '').str.replace(',', '').astype('float')
```

```python
df_cleaned['discount_percentage'] = df_cleaned['discount_percentage'].str.replace('%', '').astype('float')
```

```python
df_cleaned['rating'].unique()
```

```python
df_cleaned.loc[df.product_id == "B08L12N5H1", 'rating'] = "3.9"
```

```python
df_cleaned['rating'] = df_cleaned['rating'].astype('float')
```

```python
# Questions to solve:

# top 20 highest rating item
# Top 10 rating count
# who has the highest spending
# Top 10 discounted price
# Top 10 discounted percentage
```

```python
# top 20 highest rating item

result = df_cleaned.groupby('new_category')['rating'].mean()
result_df= result.to_frame()
result_df1 = result_df.reset_index()
new_result = result_df1.sort_values(by='rating', ascending=False)
new_df_result = new_result.head(20)
new_df_result
```

```python
# Top 10 rating count

top_rating_count = df_cleaned.groupby('new_category')['rating_count'].mean()
rating_count_df= top_rating_count.to_frame()
rating_count_df1 = rating_count_df.reset_index()
rating_count_result = rating_count_df1.sort_values(by='rating_count', ascending=False)
new_rating_count_result = rating_count_result.head(20)
new_rating_count_result
```

```python
# who has the highest spending

all_user = df_cleaned.explode('user_name')

user = all_user.groupby('user_name')['discounted_price'].sum()
user_df = user.to_frame().reset_index()
top_user_df = user_df.sort_values(by='discounted_price', ascending=False).head(1)
top_user_df
```

```python
# Top 10 discounted price
price = all_user.groupby('new_category')['discounted_price'].sum()
price_df = price.to_frame().reset_index()
top_price_df = price_df.sort_values(by='discounted_price', ascending=False).head(10)
top_price_df
```

```python
# Top 10 discounted percentage
discount = all_user.groupby('new_category')['discount_percentage'].sum()
discount_df = discount.to_frame().reset_index()
top_discount_df = discount_df.sort_values(by='discount_percentage', ascending=False).head(10)
top_discount_df
```
